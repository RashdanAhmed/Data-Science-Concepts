```python
data = np.array([[4, 1, 3],
               [1, 2, 2]])

x_pos = np.arange(2)
width = 0.2

fig, ax = plt.subplots(figsize=(7, 5), dpi=300)
ax.bar(x_pos, data[:, 0], width=width, color='black', align='center', label='Black')
ax.bar(x_pos+width, data[:, 1], width=width, color='red', align='center', label='Red')
ax.bar(x_pos+2*width, data[:, 2], width=width, color='brown', align='center', label='Brown')

labels = ['Male', 'Female']
ax.set_xticks(x_pos + width)
ax.set_xticklabels(labels, size=15)

ax.legend()

ax.set_xlabel("Gender")
ax.set_ylabel("Number of participants according to gender")
ax.set_title("Hair Colour Distribution")
```

```python
labels = ['black', 'red', 'brown']
genders = ['Male', 'Female']
colours = ['r', 'y']
data = np.array([[4, 1, 3], 
                 [1, 2, 3]])

n_groups, n_colours = data.shape

x_pos = np.arange(n_groups)
print(x_pos)
width = 0.2
fig, ax = plt.subplots(figsize=(7, 5), dpi=300)

colours = ['black', 'red', 'brown']
labels = ['Black', 'Red', 'Brown']
for i in range(n_colours):
    ax.bar(x_pos + i*width, data[:, i], width=width, color=colours[i], label=labels[i], align='center')
    
ax.legend()

ax.set_xlabel("Gender")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of hair colour amongst genders")

ax.set_xticks(x_pos+width)
ax.set_xticklabels(['Male', 'Female'])
```

```python
labels = ['Black', 'Red', 'Brown']
data = np.array([[4, 1, 3], 
                 [1, 2, 2]])

x_pos = np.arange(2)
width = 0.4

n_groups, n_colours = data.shape

fig, ax = plt.subplots(figsize=(7, 5), dpi=300)

black = data[:,0]
red = data[:,1]


ax.bar(x_pos, data[:, 0], width, color='black', label='Black', align='center')
ax.bar(x_pos, data[:, 1], width,  bottom=black, color='red', label='Red', align='center')
ax.bar(x_pos, data[:, 2], width, color='brown', bottom=black+red, label='Brown', align='center')
ax.legend(loc=0)
ax.set_xlabel("Gender")
ax.set_ylabel("frequency")
ax.set_title("Hair Colour Distribution among gender")

ax.set_xticks(x_pos)
ax.set_xticklabels(['Male', 'Female'])
```

```python
labels = ['Black', 'Red', 'Brown']
male_freq = [4, 1, 3]
female_freq = [1, 2, 2]
x_pos = np.arange(len(labels))

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

# Your code here
# 1. using ax[0].bar and ax[1].bar to plot bar charts for the first and the second subplots
w1 = 0.2
w2 = 0.4

ax[0].bar(x_pos, male_freq, width=w1, align='center', label='Male', color='r')
ax[0].bar(x_pos+width, female_freq, width=w1, align='center', label='Female', color='y')
ax[1].bar(x_pos, male_freq, width=w2, align='center', label='Male', color='r')
ax[1].bar(x_pos, female_freq, width=w2, bottom=male_freq, align='center', label='Female', color='y')

# 2. set xlabel, ylabel and title for each subplot
ax[0].set_xlabel("Hair Colour")
ax[0].set_ylabel("Frequency")
ax[0].set_title("Distribution of gender amongst hair colours")
ax[1].set_xlabel("Hair Colour")
ax[1].set_ylabel("Frequency")
ax[1].set_title("Distribution of gender amongst hair colours")

# 3. set xtick and xticklabels for each subplot
ax[0].set_xticks(x_pos+width/2)
ax[0].set_xticklabels(labels)
ax[1].set_xticks(x_pos)
ax[1].set_xticklabels(labels)

# 4. set legen for each subplot
ax[0].legend()
ax[1].legend(loc=0)
```

```python
freq = np.array([0, 2, 1, 5, 7])
np.cumsum(freq)
```

```python
freq_cumsum = np.cumsum(freq)
labels = ['Never', 'Rarely', 'Sometimes', 'Most-times', 'Always' ]
x_pos = np.arange(len(freq))
fig, ax = plt.subplots()
ax.step(x_pos, freq_cumsum, where='mid')

ax.set_xlabel("Fastening seatbelt behaviour")
ax.set_ylabel("Cumulative frequency")
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
```
